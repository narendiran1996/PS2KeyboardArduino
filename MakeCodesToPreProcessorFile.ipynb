```python
with open('PrintableMakeCodes.csv','r') as fil:
    gg = fil.readlines()
    fil.close()

scanList = [255 for i in range(256)]
for i in range(len(gg)):
    if i != 44:
        gv = gg[i].strip('\n').split(',')
        if 65 <= ord(gv[0]) and ord(gv[0]) <= 90: # alphabets
            scanList[int('0x'+gv[1], 16)] = ord(gv[0]) + 32
        else:
            scanList[int('0x'+gv[1], 16)] = ord(gv[0])
        # print(gv)
    else: # handling ","
        gv = gg[i].strip('\n')
        scanList[int('0x'+gv[4:], 16)] = ord(',')
print(scanList)
```

```python
st = 'uint8_t scanCodes[256] = { ';
for i in range(255):
    st = st + str(scanList[i]) +', '
st = st[:-2] + ' };'
print(st)
```
