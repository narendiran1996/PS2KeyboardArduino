```python
with open('PrintabelMakeCodesWithShifted.csv','r') as fil:
    gg = fil.readlines()
    fil.close()

scanList = [[0,0] for i in range(256)]
for i in range(len(gg)):
    if i != 44:
        gv = gg[i].strip('\n').split(',')
        scanList[int('0x'+gv[2], 16)] = [ord(gv[0]), ord(gv[1])]
    else: # handling ","
        gv = gg[i].strip('\n')
        scanList[int('0x'+gv[6:], 16)] = [ord(','), ord('<')]
        # print(gv[6:])
print(scanList)
```

```python
st = 'uint16_t scanCodes_ascii_withShifted[256] = { ';
for i in range(255):
    st = st + str((scanList[i][1]*256)+scanList[i][0]) +', '
st = st[:-2] + ' };'
print(st)
```
